```python
# reading worldcup.json file

import json
import pandas as pd

with open('./worldcup.json', 'r') as file:
    data = json.load(file)
    
data.keys()
```

```python
matches = data['matches']
len(matches)
```

```python
original_match_df = pd.json_normalize(matches)
print(list(original_match_df.columns))
filtered_match_df = original_match_df.loc[:, ['tournament_id', 'stage_name', 'match_date', 'match_time', 
                                              'stadium_name', 'city_name', 'country_name', 'home_team_name', 'away_team_name',
                                             'home_team_score', 'away_team_score', 'home_team_score_penalties', 'away_team_score_penalties',
                                             'result', 'home_team_win', 'away_team_win', 'draw']]
filtered_match_df
```
